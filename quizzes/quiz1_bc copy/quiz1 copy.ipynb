(a) Create two random variables ...


- Call them "vec_x" and "vec_z"
- Generate "vec_x" from a uniform distribution with (low = 10, high = 20, size = 1000)
- Generate "vec_z" from a uniform distribution with (low = -2, high = 5, size = 1000)
- Plot a histogram of each variable
    - Label the axes and title
    - Each graph should have a different title
 
To get full points, plot the histograms on a grid (See Lecture 6)

````python
# Write your answer here:
vec_x = np.random.uniform(low = 10, high = 20, size = 1000)
vec_z = np.random.uniform(low = -2, high = 5, size = 1000)

fig, list_subfig = plt.subplots(1, 2,figsize = (6,3))

plt.tight_layout()

# First Figure
list_subfig[0].hist(x = vec_x)
list_subfig[0].set_title("Uniform Distribution of vec_x")
list_subfig[0].set_xlabel("Value")
list_subfig[0].set_ylabel("Frequency")

# Second Figure
list_subfig[1].hist(x = vec_z)
list_subfig[1].set_title("Uniform Distribution of vec_z")
list_subfig[1].set_xlabel("Value")
list_subfig[1].set_ylabel("Frequency")
````
